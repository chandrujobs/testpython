```python

print("=== WELCOME TO Indigo AIRWAYS ===\n")

# Collect booking details
name = input("Enter passenger name: ")
travel_date = input("Enter date of travel (DD-MM-YYYY): ")
source = input("Enter departure city: ")
destination = input("Enter destination city: ")
passengers = int(input("Enter number of passengers: "))

# Fixed price per passenger
ticket_price = 4500.0  # in â‚¹

# Calculate subtotal
subtotal = passengers * ticket_price

# Calculate GST (5%)
gst_rate = 0.05
gst_amount = subtotal * gst_rate

# Calculate total fare (no discount)
total_amount = subtotal + gst_amount

# Print booking summary
print("\n----- TICKET SUMMARY -----")
print("Passenger Name   :", name)
print("Travel Date      :", travel_date)
print("From             :", source)
print("To               :", destination)
print("No. of Passengers:", passengers)
print("Price per Ticket : â‚¹", ticket_price)
print("------------------------------")
print("Subtotal         : â‚¹", subtotal)
print("GST (5%)         : â‚¹", gst_amount)
print("------------------------------")
print("Total Fare       : â‚¹", total_amount)
print("------------------------------")
print("Booking Confirmed! Have a safe flight with Indigo AIRWAYS")
```
