What is Data Type?

Every value in python has a type - it tells python what kind of data it is and how to handle it.

ex : You can add "Two numbers" but you can't add two names.

Main Data types :

int => Integer (Whole numbers)
float => Decimal numbers
str = String (Text)
Bool = Boolean (True / False)

```python
## Integer - Whole numbers without a decimal point

student_count = 50
age = 30
year = 2024

print(student_count)
print(age)

print(type(age))
print(type(student_count))
print(type(year))
```

```python
## Float - Used when precision is need - Includes decimal points

Price = 19.99
Height = 5.9
weight = 70.5
tax_rate = 0.07

print(Price)
print(type(weight))
```

```python
# String - Sequence of characters enclosed in single or double quotes ( "", '') used for letters, words, sentences and characters, alhaphanumeric data

name = "Heena"
student_name = 'komal' 
address = '123 Main St, Cityville'
email = "student@gmail.com"
city = "New York"

print(name)
print(student_name)
print(type(email))
print(type(address))
```

```python
# Boolean - Represents one of two values: True or False

is_student = True

has_license = True

is_adult = False

print(is_student)

print(type(has_license))

print(is_adult)


# Equal t0 >
# Less than  <
# Greater than  >=
# Less than or equal to <=
# Not equal to !=
```
